```python
import pandas as pd
import numpy
from IPython.display import HTML, display
from sklearn import feature_selection
import zhishu
import common

# 只取10年前的数据
from_date = "2016-01-14"
data_dir = "data-指数-260114"

pd.set_option("display.max_columns", None,
              "display.float_format",lambda x : '%.4f' % x)

# 宽基
zhishu_list = [
    "SH000016-上证50",
    "SH000300-沪深300",
    "SZ399006-创业板指",
    "SH000510-中证A500",
]
kuanji_df = zhishu.calc_dim_correlations(data_dir, zhishu_list, from_date)
common.print_title("宽基 - 维度-价格 相关系数")
display(kuanji_df)


# 消费民生
zhishu_list = [
    'SH000932-中证消费',
    # 'CSIH30533_中国互联网50',
    'SZ399986-中证银行',
    'SZ399989-中证医疗',
    'CSI930726-中证生物医药',
]
minsheng_df = zhishu.calc_dim_correlations(data_dir, zhishu_list, from_date)
common.print_title("消费民生 - 维度-价格 相关系数")
display(minsheng_df)

# 新科技
zhishu_list = [
    'CSI930713-人工智能',
    'CSI000941-新能源',
    'CSIH30184-半导体',
    'SZ399976-CS新能源车',
    'CSIH30590-机器人',
    'CSI931798-光伏龙头'
]
tech_df = zhishu.calc_dim_correlations(data_dir, zhishu_list, from_date)
common.print_title("新科技 - 维度-价格 相关系数")
display(tech_df)

# 基础工业
zhishu_list = [
    "SH000928-中证能源",
    "SZ399440-国证钢铁",
    "CSI930632-CS稀金属",
    "SH000819-有色金属",
    "CSIH30199-电力指数"
]
industry_df = zhishu.calc_dim_correlations(data_dir, zhishu_list, from_date)
common.print_title("基础工业 - 维度-价格 相关系数")
display(industry_df)

# 国防军工
zhishu_list = [
    "SZ399967-中证军工", 
    "CSI931066-军工龙头",
]
military_df = zhishu.calc_dim_correlations(data_dir, zhishu_list, from_date)
common.print_title("国防军工 - 维度-价格 相关系数")
display(military_df)

# 夕阳
zhishu_list = [
    'CSI931009-建筑材料',
    'SZ399995-基建工程',
]
hopeless_df = zhishu.calc_dim_correlations(data_dir, zhishu_list, from_date)
common.print_title("夕阳产业 - 维度-价格 相关系数")
display(hopeless_df)
```

```python
kuanji_dim_df = common.get_highest_correlation_dim(kuanji_df)
common.print_title("宽基 - 最相关系数")
display(kuanji_dim_df)

mingsheng_dim_df = common.get_highest_correlation_dim(minsheng_df)
common.print_title("消费民生 - 最相关系数")
display(mingsheng_dim_df)

tech_dim_df = common.get_highest_correlation_dim(tech_df)
common.print_title("新科技 - 最相关系数")
display(tech_dim_df)

industry_dim_df = common.get_highest_correlation_dim(industry_df)
common.print_title("基础工业 - 最相关系数")
display(industry_dim_df)

military_dim_df = common.get_highest_correlation_dim(military_df)
common.print_title("国防军工 - 最相关系数")
display(military_dim_df)

hopless_dim_df = common.get_highest_correlation_dim(hopeless_df)
common.print_title("夕阳产业 - 最相关系数")
display(hopless_dim_df)
```
